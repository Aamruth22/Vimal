```python
import cv2 as cv
import os
import pytesseract
from typing import OrderedDict
import matplotlib.pyplot as plt
from fuzzywuzzy import fuzz
from fuzzywuzzy import process

path = '/home/vimalkumar/Downloads/cells'
for im in os.listdir(path):
    img = cv.imread(os.path.join(path,im),0)
    image_data = pytesseract.image_to_data(img, output_type = pytesseract.Output.DICT)

    key_vals = ['INVOICE','DESCRIPTION','vat','details','Materials','Debit','Unit']
    to_find = key_vals
    keys = []
    values = []

    for key in key_vals:
        to_find = key.lower()
        found_val = ""
        for i in range(len(image_data['text'])):
            if to_find in image_data['text'][i].lower():
                x,y,w,h = image_data['left'][i], image_data['top'][i], image_data['width'][i], image_data['height'][i]
                if to_find == "INVOICE":
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)
                elif to_find == "DESCRIPTION":
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)
                elif to_find=="vat":
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)
                elif to_find=='DETAILS':
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)
                elif to_find=='Materials':
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)
                elif to_find=='Unit':
                    zoning_area = (x-5,y-20,x+w+100,h+y+10)    
                elif to_find=='debit':
                    zoning_area = (x-5,y-20,x+w+100,h+y+10) 
                else:
                    zoning_area = (x-10,y-10,x+w+250,h+y+10)
                cv.rectangle(img, (zoning_area[0], zoning_area[1]), (zoning_area[2],zoning_area[3]), (0,255,0), 2)
                break


        for i in range(len(image_data['text'])):
            x,y,w,h = image_data['left'][i], image_data['top'][i], image_data['width'][i], image_data['height'][i]
            if x>=zoning_area[0] and y >= zoning_area[1] and x+w<=zoning_area[2] and h+y<zoning_area[3]:
                found_val += image_data['text'][i]
                #print(f"{to_find} = {found_val}")



        #print(f"{to_find} = {found_val}")
        keys.append(to_find)
        values.append(found_val)
    #cv.imshow("ABC", img)
    #cv.waitKey(0)
    #cv.imwrite("ABC.Jpeg",img)
    plt.figure(figsize=(12,12))
    plt.imshow(img)

    dic = dict(zip(keys,values))

    if fuzz.partial_ratio('invoice',dic['invoice'].lower())>95 or fuzz.partial_ratio('debit',dic['debit'].lower())>95:
        th_invoice = 50
    else:
        th_invoice = 0 
    if fuzz.partial_ratio('description',dic['description'].lower())>95 or fuzz.partial_ratio('details',dic['details'].lower())>95 or fuzz.partial_ratio('materials',dic['materials'].lower())>95:
        th_des = 20
    else:
        th_des = 0
    if fuzz.partial_ratio('vat',dic['vat'].lower())>95:
        th_vat = 20
    else:
        th_vat = 0
    if fuzz.partial_ratio('unit',dic['unit'].lower())>95:
        th_unit = 10
    else:
        th_unit = 0
    #print(th_invoice,th_des,th_vat,th_unit)    

    threshold = th_invoice+th_des+th_vat+th_unit
    print(threshold)

    if threshold>=80:
        print(im , ':' , 'INVOICE')
    else:
        print(im , ':' , 'Unclassified')

```
