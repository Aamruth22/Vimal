```python
import cv2 as cv
import pytesseract

img = cv.imread('0001.tif')
```

```python
data = pytesseract.image_to_data(img, output_type=pytesseract.Output.DICT)
print(data["text"])
to_find = ("Violation")
```

```python

for i in range(len(data['text'])):
    if to_find == "Violation":
        x,y,w,h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        zoning_area = (x-10,y-20,x+w+100,h+y+10)
    
    else:
        x,y,w,h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
        zoning_area = (x-10,y-20,x+w+100,h+y+10)
        break
cv.rectangle(img, (x-50,y-50), (x+w+100, y+h+50), (0,234,0), 2)
```

```python
for i in range(len(data['text'])):
    x,y,w,h = data['left'][i], data['top'][i], data['width'][i], data['height'][i]
    if x>=zoning_area[0] and y >= zoning_area[1] and x+w<=zoning_area[2] and h+y<zoning_area[3]:
        print(data['text'][i])

cv.imwrite("IMAGE.png",img)
```
