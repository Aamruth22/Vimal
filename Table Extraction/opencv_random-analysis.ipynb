```python
import numpy
import cv2
import matplotlib
import os
import shutil ,random
import matplotlib.pyplot as plt
def main():
    inPath ="/home/vimal/Documents/Table layout extraction/table_detection input/1st page_jpeg/04152020"
    outPath ="/home/vimal/Documents/Table layout extraction/open-CV method/Random Analysis"
    #filenames = random.sample(os.listdir(dirpath), 10)
    for imagePath in random.sample(os.listdir(inPath), 10):
        inputPath = os.path.join(inPath, imagePath)
        img = cv2.imread(inputPath)
        newpath = os.path.splitext(imagePath)[0] + '.jpeg'
        fullOutPath = os.path.join(outPath, ''+newpath)
        matplotlib.image.imsave(fullOutPath,img)
        print(fullOutPath)
if __name__ == '__main__': 
    main()        
```

```python
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import cv2
import matplotlib
image_path='/home/vimal/Documents/Table layout extraction/open-CV method/Random Analysis/Images/0944664000.jpeg'
image=cv2.imread(image_path)
#plt.figure(figsize=(8,10))
#plt.imshow(image)

gray_scale=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
th1,img_bin = cv2.threshold(gray_scale,150,225,cv2.THRESH_BINARY)
img_bin=~img_bin



line_min_width = 15
kernal_h = np.ones((1,line_min_width), np.uint8)
kernal_v = np.ones((line_min_width,1), np.uint8)

img_bin_h = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, kernal_h)



img_bin_v = cv2.morphologyEx(img_bin, cv2.MORPH_OPEN, kernal_v)

img_bin_final=img_bin_h|img_bin_v

_, labels, stats,_ = cv2.connectedComponentsWithStats(~img_bin_final, connectivity=8, ltype=cv2.CV_32S)

fig = plt.figure(figsize=(20,20))
plt.tight_layout()
axes1 = fig.add_subplot(1,2,1)
axes2 = fig.add_subplot(1,2,2)
axes1.imshow(image)
#axes1.set_title('0384265000',fontsize = 20)
axes2.imshow(labels)
```
